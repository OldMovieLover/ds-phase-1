**4.** Мы являемся владельцами розничного магазина и хотим оптимизировать нашу обслуживающую систему, чтобы обеспечить удобство покупателей и эффективность работы магазина. * Что это за распределение?

Распределение Пуассона (число событий, произошедших за фиксированное время)

* Найти вероятность того, что за 2 минуты в магазин зайдет не более 1 человека.

р = 0.0174

* Визуализируй это распределение с вычисленным параметром с помощью бибилиотеки `scipy`

**7.** Рост мальчиков возрастной группы 15 лет - нормально распределённая случайная величина $X$ с параметрами $\mu=161$ см и $\sigma=4$ см.
* Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы?

0.2144

```python
mu = 161 
sigma = 4
a = (152 - mu) / sigma
a1 = (158 - mu) / sigma
p = st.norm.cdf(a1) - st.norm.cdf(a)
p
```

* Сформулируй правило трёх сигм для случайной величины $X\sim\mathcal{N}$.

* 1 сигма (sigma = 4) (mu +- sigma) (68%): рост от 157 до 165;
* 2 сигмы (sigma * 2 = 8) -//- (95%): рост от 153 до 169;
* 3 сигмы (sigma * 3 = 12) -//- (99.7%): рост от 149 до 173.

* Визуализируй следующие элементы: 
    * кривую плотности вероятности для указанных параметров и распределения
    * залей выдели цветом нужный фрагмет диаграммы (примерно это должно выглядеть [так](https://i.stack.imgur.com/c8F3F.png), только с твоими параметрами)

```python
# code
mu = 161
sigma = 4

x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
y = st.norm.pdf(x, mu, sigma)  
lower_bound = 152  
upper_bound = 158 

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Нормальное распределение', color='blue')
plt.fill_between(x, y, where=(x >= lower_bound) & (x <= upper_bound), color='lightgreen', alpha=0.5, label='Рост от 152 до 158 см')
plt.title('Кривая плотности вероятности роста мальчиков (15 лет)')
plt.xlabel('Рост (см)')
plt.ylabel('Плотность вероятности')
plt.axvline(mu, color='red', linestyle='--', label='Средний рост (161 см)')
plt.axvline(lower_bound, color='orange', linestyle='--', label='Нижний предел (152 см)')
plt.axvline(upper_bound, color='orange', linestyle='--', label='Верхний предел (158 см)')
plt.legend()
plt.grid()
plt.show()
```

❓От чего зависит форма кривой нормального распределения? 

* От среднего mu: Сдвигает кривую влево или вправо.
* И стандартного отклонения sigma: Определяет ширину кривой; большая sigma делает кривую широкой, а меньшее — узкой.
